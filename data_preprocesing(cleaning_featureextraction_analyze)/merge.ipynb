```python
output = pd.merge(output,weather, on=['data_block_id'], how='inner')
```

```python
import pandas as pd

# Assuming 'production' and 'weather' are your dataframes

# Convert data_block_id to the same data type to ensure matching works correctly
output['data_block_id'] = output['data_block_id'].astype(float)
weather['data_block_id'] = weather['data_block_id'].astype(float)

# Initialize an empty list to hold merged data
merged_data = []

# Loop through each row in the production dataframe
for idx, prod_row in output.iterrows():
    # Find matching row in weather dataframe based on data_block_id
    match = weather[weather['data_block_id'] == prod_row['data_block_id']]

    # If there's a match, merge the two rows
    if not match.empty:
        for _, weather_row in match.iterrows():
            merged_row = {**prod_row.to_dict(), **weather_row.to_dict()}
            merged_data.append(merged_row)

# Convert the list of dictionaries to a dataframe
merged_df = pd.DataFrame(merged_data)

# Display the information of the merged dataframe
print(merged_df.info())
```
